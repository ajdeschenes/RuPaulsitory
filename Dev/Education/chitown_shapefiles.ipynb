```python
from descartes import PolygonPatch
import shapefile
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
sf=shapefile.Reader('Resources/boundaries_community_areas/geo_export_e1fd9f41-7481-47c6-a15f-cc8ae6d1bee9')
fig = plt.figure() 
ax = fig.gca() 
for poly in sf.shapes():
    poly_geo=poly.__geo_interface__
    ax.add_patch(PolygonPatch(poly_geo, fc='#ffffff', ec='#000000', alpha=0.5, zorder=2))

ax.axis('scaled')
plt.show()
```
